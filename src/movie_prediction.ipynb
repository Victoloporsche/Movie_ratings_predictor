```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn .model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelBinarizer
```

```python
#load the movies and rating data
movies = pd.read_csv('./input/movies_metadata.csv')
ratings = pd.read_csv('./input/ratings_small.csv')
```

```python
#Utilize the most important features from the movie metadata
movies_df = movies[['id', 'revenue', 'runtime', 'status', 'vote_average']]
```

```python
#replace missing values to Unkown
movies_df['status'] = movies_df[['status']].fillna('Unknown')
```

```python
#drop the timestamp column from ratings
ratings_df = ratings.drop(['timestamp'], axis=1)
```

```python
#convert column from object to integer
def to_numeric(data, column):
    data[column] = pd.to_numeric(data[column], errors='coerce', downcast='integer')
    data = data.fillna(0)
    data[column] = data[column].astype(int)
    return data
```

```python
# Id column converted to type integer
movies_df= to_numeric(movies_df, column='id')
```

```python
# merge the movies and rating dataframe to one
movies_ratings_df = pd.merge(ratings_df, movies_df, left_on='movieId', right_on='id')
```

```python
# remove the id column from the dataframe
movies_ratings_df = movies_ratings_df.drop(['id'], axis=1)
```

```python
# plot of ratings and their frequencies in the data
movies_ratings_df['rating'].value_counts().plot(kind='bar')
plt.xlabel('Ratings')
plt.ylabel('Frequencies')
plt.show()
```

```python
top_10_rated_movies = movies_ratings_df.drop_duplicates(subset='movieId').sort_values('rating', ascending = False).head(10)
```

```python
least_10_rated_movies = movies_ratings_df.drop_duplicates(subset='movieId').sort_values('rating', ascending = True).head(10)
```

```python
movies_features = movies_ratings_df[['userId', 'movieId', 'revenue', 'runtime', 'status', 'vote_average']]
```

```python
for feature in movies_features:
    data = top_10_rated_movies.copy()
    data.groupby(feature)['rating'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('rating')
    plt.title(feature)
    plt.show()
```

```python
for feature in movies_features:
    data = least_10_rated_movies.copy()
    data.groupby(feature)['rating'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('rating')
    plt.title(feature)
    plt.show()
```

```python
# encode categorical features
def encode_categorical(data):
    labelEncoder=LabelEncoder()
    column = ['status']
    for column in data:
        labelEncoder.fit(data[column])
        data[column]=labelEncoder.transform(data[column])
    return data

#scale the features
def scale_data(data):
    scaling_feature=[feature for feature in data.columns if feature not in ['rating']]
    scaler=MinMaxScaler()
    scaler.fit(data[scaling_feature])
    scaler.transform(data[scaling_feature])
    data = pd.concat([data[['rating']].reset_index(drop=True), 
                 pd.DataFrame(scaler.transform(data[scaling_feature]), columns=scaling_feature)], axis=1)
    return data
    
    
```

```python
#the categorical features are encoded with meaningful integers
movies_ratings_encoded_df = encode_categorical(movies_ratings_df)
```

```python
#the features are scaled
movies_ratings_encoded_scaled_df = scale_data(movies_ratings_encoded_df)
```

```python
#selecting the features needed to train the model
x_data = movies_ratings_encoded_scaled_df[['userId', 'movieId', 'revenue','runtime', 'status', 'vote_average']]
```

```python
#the ratings for every movie
y_data = movies_ratings_encoded_scaled_df[['rating']]
```

```python
#split the dataset into train and test
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size= 0.2, random_state=42)
```

```python
#function which perform training and prediction on the model
def train_test_model(model_name, xtrain, ytrain, xtest, ytest):
    model = model_name
    model.fit(x_train, y_train)
    prediction = model.predict(x_test)
    return prediction
```

```python
# Utilizing random forest classfier
rf_model = RandomForestClassifier()
y_predict = train_test_model(rf_model, x_train, y_train, x_test, y_test)
```

```python
#function to check the accuracy of the model
def multiclass_roc_auc_score(y_test, y_pred, average="macro"):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    return roc_auc_score(y_test, y_pred, average=average)
```

```python
#prints the accuracy of the prediction model
accuracy = multiclass_roc_auc_score(y_test, y_predict)
print('The accuracy of the prediction model of the test set is:', accuracy)
```
